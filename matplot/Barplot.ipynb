```python
import matplotlib.pyplot as plt

y = [98,56,98,89,76,49]
x = ["part_1","part_2","part_3","part_4","part_5","part_6"]
colors = ["red","yellow","pink","green","blue","purple"]
plt.bar(x , y, color = colors)
plt.xlabel("Parts of Harry Potter",fontsize = 15)
plt.ylabel("Popularity", fontsize = 15)
plt.title("Popularity Against Harry Porter Parts", fontsize = 18, fontweight = "bold")
plt.savefig("Bar.png")#used to save charts in img fromat

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

cars = pd.read_csv("cars.csv")
df = pd.DataFrame(cars)
print(df)
grouped_by = df.groupby("Model")["Horsepower"].sum()
print(grouped_by)
plt.bar(grouped_by.index,grouped_by.values)
plt.xlabel("Modle")
plt.ylabel("HorsePower")
plt.show()
```
