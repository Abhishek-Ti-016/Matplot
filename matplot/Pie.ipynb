```python
import matplotlib.pyplot as plt
import pandas as pd

brands = ["Oneplus","Samsung","Nokia","Apple","Vivo"]
x = [12,38,7,40,10]
ex = [0,0.05,0,0,0]
plt.pie(x, labels = brands,explode = ex,shadow = True,autopct = "%.f",startangle = 45)
plt.show()
```
