```python
import matplotlib.pyplot as plt

x = ["day1","day2","day3","day4","day5"]
y = [300,400,450,681,785]
y1 = [350,200,200,800,964]

plt.plot(x,y, marker = "D",ls = "--",label = "Week 1")
plt.plot(x,y1, marker = "D", color = "red", label = "Week 2")#alpha = 0 - 1 used to reduce line transprancy 
plt.legend()
plt.show()
```
